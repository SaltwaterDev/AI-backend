```python
import pandas as pd
import numpy as np
```

```python
def mean_reciprocal_rank(labels, predictions):
    assert len(labels) == len(predictions)
    
    total_score = 0
    rank = 1
    true = labels['sentence2'].iloc[0]
    for i in range(len(labels)):    
        if true != predictions['sentence2'].iloc[i]:
            rank += 1
        else:
            reciprocal_rank = 1/rank
            total_score += reciprocal_rank
            rank = 1
            
        if i+1 < len(labels) and labels.iloc[i].sentence1 != labels.iloc[i+1].sentence1:
            true = labels['sentence2'].iloc[i+1]
           
                
    rnn = total_score/len(labels)
    return rnn
```

```python
#orgin_df_test = pd.read_csv('df_reddit_test.csv')
orgin_df_test = pd.read_csv('df_test.csv')
df_test = orgin_df_test.sort_values(['sentence1', 'score'], ascending=False)
df_test
```

```python
path_to_output_file = f'./results/all_log_MSELoss()_output.txt'  # path to the file with prediction probabilities
probs_test = pd.read_csv(path_to_output_file, header=None)[0]  # prediction probabilities
predict_result = orgin_df_test.copy()
predict_result['score'] = probs_test
predict_result = predict_result.sort_values(['sentence1', 'score'], ascending=False)
predict_result
```

```python
mean_reciprocal_rank(df_test, predict_result)
```
